```python
import pandas as pd

#Read each CSV file
pm10_df = pd.read_csv('income_pm10_merged.csv')
pm25_df = pd.read_csv('income_pm25_merged.csv')
ozone_df = pd.read_csv('income_ozone_merged.csv')
nitrogen_df = pd.read_csv('income_nitrogen_merged.csv')
co_df = pd.read_csv('income_carbon_merged.csv')

#Merge all dataframes on city_index, city, month, and median_income_2020
master_df = pd.merge(pm10_df, pm25_df, on=['city_index', 'city', 'month', 'median_income_2020'], how='outer')
master_df = pd.merge(master_df, ozone_df, on=['city_index', 'city', 'month', 'median_income_2020'], how='outer')
master_df = pd.merge(master_df, nitrogen_df, on=['city_index', 'city', 'month', 'median_income_2020'], how='outer')
master_df = pd.merge(master_df, co_df, on=['city_index', 'city', 'month', 'median_income_2020'], how='outer')

#Save the master dataframe to a CSV file
master_df.to_csv('city_master_list.csv', index=False)

master_df
```
